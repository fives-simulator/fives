## Plot of the empirical bandwidth degradation model

```python
import seaborn as sns
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
flights = sns.load_dataset("flights")
flights.tail()
```

```python
bw_max = 1000

x = np.arange(1, 20, 0.1)

data = {"x":[], "y": [], "c": []}

def model_1(x, c, max_bw=1):
    return max_bw / np.log(c * x)

def model_2(x, c, max_bw=1):
    return max_bw / np.sqrt(c * x) 

def model_3(x, c, max_bw=1):
    return max_bw / (c * x) 


for c in np.arange(1, 20, 1):
    y = [model_2(x_i, c) for x_i in x]
    data["x"].extend(x)
    data["y"].extend(y)
    data["c"].extend([str(c) for _ in x])

df = pd.DataFrame(data)
```

```python
g = sns.lineplot(x="x", y="y", hue="c", data=df)
# g.set(yscale='log')
```

```python
x = np.linspace(1, 1000, 1000)

max_nodes = 1000
beta = 10
gamma = 0.01
y = [max_nodes * np.exp((-beta) * np.exp((-gamma) * x_i)) for x_i in x] 
data = {"x":x, "y": y}
df = pd.DataFrame(data)
```

```python
g = sns.lineplot(x="x", y="y", data=df)
```
